### Subset dataframe by a range of date

### this can be done by using boolean expression wiht date time index as
- df['begin_index_date' : 'end_index_date']

```python
#display data from 2011-1-1 to 2012-1-1
data_2011_to_12=data['2011-1-1':'2012-1-1'].head()
```

```python
data_2011_to_12.tail()
```

```python
#plot the subset data from 2011 to 2013
fig, ax =plt.subplots(figsize=(20,10))
ax.bar(data_2011_to_12.index.values,
      data_2011_to_12['DAILY_PRECIP'],
      color='red')
#set the title of axies
ax.set(xlabel='Date',
      ylabel='Daily Precipation',
      title="Daily Precipation-2011 to 2013")

# Rotate tick marks on x-axis
plt.setp(ax.get_xticklabels(), rotation=45)
plt.show()
```
